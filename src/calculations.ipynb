```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
print(X)
```

```python
def f(l, x):
    return l*x*(1-x)


xn = [0]
l = 0.7
list = []
while l <= 4:
    xn = np.random.random()
    for i in range(2000):
        xn = f(l, xn)
    list.append((xn, l))
    l += 0.001
print(list)
fig1 = plt.figure(figsize = (7,7))
ax = fig1.add_subplot(111)
ax.scatter([i[1] for i in list], [i[0] for i in list], label="name")
fig1.show()
```
